# Hypothesis Testing

**1. Cost efficiency**

Null hypothesis(H0): There is no significant difference in cost per kilometre between the yellow and pink cab

Alternate hypothesis(H1): There is a significant difference incost per kilometre between the pink and yellow cab.

```python
# Testing the cost efficiency hypothesis using T_statistic and p_value

YellowCab_cost_efficiency = final_df[final_df['Company'] == 'Yellow Cab']['Cost per kilometer']
PinkCab_cost_efficiency = final_df[final_df['Company'] == 'Pink Cab']['Cost per kilometer']

T_statistic, p_value = stats.ttest_ind(YellowCab_cost_efficiency, PinkCab_cost_efficiency)
print(f'Hypothesis cost efficiency: T_statistic={T_statistic:.4f}, p_value={p_value}')

```

The p_value is (0.0)  which is less than the typical chosen significance levle(0.05) therefore, we reject the null hypothesis. There is a significant difference in cost per kilometer in every company(pink and yellow cab)

**2. Companies' Users Income Level**

Null hypothesis(H0): The choice of cab company is not affected by the user's income level

Alternate hypothesis(H1): The choice of cab company is affected by users's income level

```python
# performing hypothesis testing for the above customer income level assumptions

YellowCab_income = final_df[final_df['Company'] == 'Yellow Cab']['Income (USD/Month)']
PinkCab_income = final_df[final_df['Company'] == 'Pink Cab']['Income (USD/Month)']

T_statistic, p_value = stats.ttest_ind(YellowCab_income, PinkCab_income)
print(f'Hypothesis income level: T_statistic={T_statistic}, p_value={p_value:.4f}')
```

The p_value is (0.6693), this is higher than the typical sigificant level(0.05) therefore, we cannot reject the null hypothesis. The choice of cab company a customer chooses is not influenced by their monthly income.

**3. Average price charged in both campanies**

H0: There is no difference in the everage price charged between the yellow and pink cab.

H1: There is a significant difference in the prces charged between the two cab companies 

To test this hypothesis we will perform a t-test comparing the average price charged for each company

```python
# performing the hypothesis Testing

YellowCab_prices = final_df[final_df['Company'] == 'Yellow Cab']['Price Charged']
PinkCab_prices = final_df[final_df['Company'] == 'Pink Cab']['Price Charged']
t_stat, p_value = stats.ttest_ind(YellowCab_prices, PinkCab_prices)

print(f'Hypothesis average price charged: t_stat = {t_stat:.4f}, p_value = {p_value}')
```

The p_value is below the typical threshold(0.05), we therefore reject the null hypothsis(H0). There is a significant difference in price charged between the two companies. Lets visualize below the city's revenue/ perfomance in every cab company.

**4. Trips Variation per city**

H0: Number of Trips does not vary across all the cities

H1: Number of trips varies across all the cities

we will perform ANOVA testing to compare the trips in all the cities

```python
trip_per_city = [final_df[final_df['City'] == city]['Transaction ID'] for city in final_df['City'].unique()]

trips_results_Anova = stats.f_oneway(*trip_per_city)

print(f'Hypothesis trip variation: F_statistic={trips_results_Anova.statistic}, p_value={trips_results_Anova.pvalue}\n')

Pvalue = 0.05

if trips_results_Anova.pvalue < Pvalue:
    print(f'The p_value is; {trips_results_Anova.pvalue:.4f}, therefore we reject the null hypotheis. The number of trips varies across all the cities')

else: 
    print(f'The p_value is; {trips_results_Anova.pvalue:.4f}, therefore we cannot reject the null hypotheis. The number of Trips does not vary across all the cities')

```

The number of trips any cab company makes varies from city to city. This proves the general knowledge where big cities are busy compared to small cities and also the diffrent number of users per city
